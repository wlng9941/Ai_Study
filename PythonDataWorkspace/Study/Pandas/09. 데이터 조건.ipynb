# 데이터 선택 (조건)
조건에 해당하는 데이터 선택

```python
import pandas as pd
df = pd.read_excel('score.xlsx', index_col='지원 번호')
df
```

```python
df['키'] >= 185 # 학생들의 키가 185 이상인지 여부 T/F
```

```python
filt =  (df['키'] >= 185)
```

```python
df[filt]
```

```python
df[~filt] #filt 역
```

```python
df[df['키'] >= 185]
```

```python
df.loc[df['키'] >= 185,'수학']
```

```python
df.loc[df['키'] >= 185,'수학':'SW특기']
```

```python
df.loc[df['키'] >= 185,['수학','SW특기']]
```

## 다양한 조선

### &그리고(and)

```python
df.loc[(df['키'] >=185) & (df['학교'] == '북산고')] # 키 185이상, 북산고학생
```

### | 또는 (or)

```python
df.loc[(df['키'] <=170) |(df['키'] >=200)]
```

### str 함수

```python
filt =df['이름'].str.startswith('송') # 송 씨 성을 가진 사람
df[filt]
```

```python
filt = df['이름'].str.contains('태') # 이름에 태 포함
df[filt]
```

```python
df[~filt] 
```

```python
langs = ['Python', 'Java']
filt = df['SW특기'].isin(langs) #SW특기가 Python, Java
df[filt]
```

```python
langs = ['python', 'java']
filt = df['SW특기'].str.lower().isin(langs) #SW특기가 Python, Java
df[filt]
```

```python
df[df['SW특기'].str.contains('Java', na = False)]
```

```python
filt = df['SW특기'].str.contains('Java', na = False)
df[filt]
```

```python
filt = df['SW특기'].str.contains('java', na = False)
df[filt]
```
